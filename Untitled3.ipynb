```python
dados = pd.read_csv("/listagem-de-servidores-terceirizados.csv", sep= ";", encoding="cp1252")

#Dando nomes as colunas
dados.columns = ['Ano', 'Funcionario', 'Cargo', 'Local de Serviço']
display(dados)
```

```python
dados.groupby(["Local de Serviço", "Cargo"]).sum().head(50)
```

```python
#Mostrando qual cargo que mais se repetem
detalhe = dados['Cargo'].value_counts()
detalhe.head(50)
```

```python
detalhe2 = dados[['Cargo', 'Local de Serviço']].value_counts()
detalhe2.head(50)
```

```python
empresas = dados['Local de Serviço'].value_counts()
display(empresas)
```
